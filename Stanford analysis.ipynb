```python
window = 5
df["mean_pts"] = df.groupby("Name")["PTS"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_orb"] = df.groupby("Name")["ORB"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_drb"] = df.groupby("Name")["DRB"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_ast"] = df.groupby("Name")["AST"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_stl"] = df.groupby("Name")["STL"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_blk"] = df.groupby("Name")["BLK"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_+-"] = df.groupby("Name")["+-"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_fga"] = df.groupby("Name")["FGA"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_fg"] = df.groupby("Name")["FG"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_fta"] = df.groupby("Name")["FTA"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_ft"] = df.groupby("Name")["FT"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_tov"] = df.groupby("Name")["TOV"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
df["mean_pf"] = df.groupby("Name")["PF"].expanding(window).agg(lambda rows: rows[:len(rows)-1].mean()).reset_index(level = 0,drop = True)
```

```python
df["Fantasy_predict"] = 0.15 * df["mean_+-"] + df["mean_pts"] - 0.5*(df["mean_fga"] - df["mean_fg"] + df["mean_fta"] - df["mean_ft"]) + 0.75*df["mean_drb"] + df["mean_orb"] + df["mean_ast"] + 2*df["mean_stl"] + 2*df["mean_blk"] - 0.75*df["mean_tov"] - 0.5*df["mean_pf"]
```

```python
df = pd.read_csv('allData.csv', sep=',',low_memory=False)
```

```python
def weighted_stats(rows):
    rows = rows.reset_index(drop=True)
    return (rows[:window-1]*(np.array(rows.index.tolist()[:window-1])+1)).sum()
```

```python
window = 11
df_mean = df.groupby("Name")[['FG', 'FGA','FT', 'FTA', 'ORB', 'DRB','AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', '+-']].rolling(window).agg(weighted_stats).reset_index(level = 0,drop = True)
```

```python
df_mean = df_mean.rename({'FG':'mean_FG', 'FGA':'mean_FGA','FT':'mean_FT', 'FTA':'mean_FTA', 'ORB':'mean_ORB', 'DRB':'mean_DRB','AST':'mean_AST', 'STL':'mean_STL', 'BLK':'mean_BLK', 'TOV':'mean_TOV', 'PF':'mean_PF', 'PTS':'mean_PTS', '+-':'mean_+-'},axis=1)
```

```python
df_weighted_stats = pd.concat([df, df_mean], axis=1)
df_weighted_stats.columns
```

```python
df_weighted_stats["prev_date"] = df_weighted_stats.groupby("Name")["Date"].shift(1)
df_weighted_stats = df_weighted_stats.dropna()
df_weighted_stats["prev_date"] = df_weighted_stats["prev_date"].astype(int)
df_weighted_stats 
```

```python
df_weighted_stats["rest_days"] = df_weighted_stats.apply(lambda row: (datetime.strptime(str(row["Date"]), "%Y%m%d") - datetime.strptime(str(row["prev_date"]),"%Y%m%d")).days,axis=1)
```

```python
df_weighted_stats
```

```python
modeleReg=LinearRegression()
X = df_weighted_stats[['mean_FG',
       'mean_FGA', 'mean_FT', 'mean_FTA', 'mean_ORB', 'mean_DRB', 'mean_AST',
       'mean_STL', 'mean_BLK', 'mean_TOV', 'mean_PF', 'mean_PTS', 'mean_+-',"rest_days"]]

Y = df_weighted_stats["Fantasy"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lmodellineaire = LinearRegression()
lmodellineaire.fit(X_train, Y_train)
```

```python
y_train_predict = lmodellineaire.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print('La performance du modèle sur la base dapprentissage')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
print('\n')
```

```python
# model evaluation for testing set
y_test_predict = lmodellineaire.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
 
print('La performance du modèle sur la base de test')
print('--------------------------------------')
print('Lerreur quadratique moyenne est {}'.format(rmse))
print('le score R2 est {}'.format(r2))
```
